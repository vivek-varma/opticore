```python
PORT = "7497"
```

```python
import nest_asyncio
nest_asyncio.apply()

from ib_async import IB, Stock

ib = IB()
ib.connect("127.0.0.1", PORT, clientId=99, timeout=10)   # <-- changed to 99
ib.reqMarketDataType(3)
stock = Stock("AAPL", "SMART", "USD")
ib.qualifyContracts(stock)

[ticker] = ib.reqTickers(stock)
ib.sleep(5)

print("marketPrice():   ", ticker.marketPrice())
print("last:            ", ticker.last)
print("close:           ", ticker.close)
print("bid / ask:       ", ticker.bid, "/", ticker.ask)
print("delayedLast:     ", getattr(ticker, 'delayedLast', None))
print("delayedClose:    ", getattr(ticker, 'delayedClose', None))
print("delayedBid / Ask:", getattr(ticker, 'delayedBid', None), "/", getattr(ticker, 'delayedAsk', None))
print("time:            ", ticker.time)

ib.disconnect()
```
